## Real Exampe - a Berlin Tile

### (latitude, longitude, zoom) -> tile

```python
zoom = 15
latitude, longitude = 52.519819, 13.297018 # a random coordinate in Berlin
```

```python
mercantile.tile(longitude, latitude, zoom)
```

### tile -> tile bounding box

```python
LngLatBbox = mercantile.bounds(mercantile.tile(longitude, latitude, 15))
print(LngLatBbox)
```

### visualize tile bounding box

```python
bbx_ox = ox.footprints.create_footprints_gdf(west=LngLatBbox.west, south=LngLatBbox.south, 
                                             east=LngLatBbox.east, north=LngLatBbox.north)

ox.footprints.plot_footprints(bbx_ox);
```

### tile -> quadkey

```python
mercantile.quadkey(mercantile.tile(longitude, latitude, zoom))
```
